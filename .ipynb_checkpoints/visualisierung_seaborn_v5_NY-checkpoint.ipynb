```python
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file = "data/12/NewYorkCity/listings_big_clean.csv" 

df = pd.read_csv(file, low_memory=False)
```

```python
df.info()
```

```python
# if "host_since" in df.columns:
#     df["host_since"] = pd.to_datetime(df["host_since"])
df["host_since"].head(10)
```

```python
# extract year from "host_since"
if "host_since" in df.columns:
    df["host_since"] = pd.to_datetime(df["host_since"])
    
df["host_since_year"] = df["host_since"].dt.year.astype("int64")
```

```python
df["host_since_year"].head(5)
```

```python
# Create a histogram of the room_type amounts

sns.histplot(data=df, x="room_type")
```

```python
# Create a histogram of the availability_365 amounts

sns.histplot(data=df, x="availability_365")
```

```python
# Create a histogram of the neighbourhood_group_cleansed amounts

sns.histplot(data=df, x="neighbourhood_group_cleansed")
```

```python
sns.scatterplot(x="neighbourhood_group_cleansed", y="price", data=df)
```

```python
sns.scatterplot(x="room_type", y="number_of_reviews", data=df)
```

```python
# customize the scatter plot

sns.scatterplot(x="number_of_reviews_ly", y="host_since_year", hue="neighbourhood_group_cleansed", data=df)

# add labels and title

plt.xlabel("number_of_reviews_ly")

plt.ylabel("host_since")

plt.title("Relationship between host_since_year and number_of_reviews_ly")

# display the plot

plt.show()
```

```python
mask = df["price"] < 5000
sns.boxplot(x="neighbourhood_group_cleansed", y="price", hue="instant_bookable", data=df[mask], palette="Set3", linewidth=1.5, fliersize=4)

# add a title, xlabel, and ylabel to the plot using Matplotlib functions

plt.title("Box Plot of Total Bill by Day and Meal Time")

plt.xlabel("Day of the Week")

plt.ylabel("Total Bill ($)")

# display the plot

plt.show()
```

```python
# Create a heatmap of the correlation between variables

corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True)

# Show the plot

plt.show()
```
