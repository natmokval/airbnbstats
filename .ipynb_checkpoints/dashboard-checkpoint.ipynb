```python
import plotly.express as px
import pandas as pd
```

```python
file = "data/12/NewYorkCity/listings_big_clean.csv" 

df = pd.read_csv(file, low_memory=False)
```

```python
df.info()
```

```python
if "host_since" in df.columns:
    df["host_since"] = pd.to_datetime(df["host_since"])

# extract year from "host_since"
df["host_since_year"] = df["host_since"].dt.year.astype("int64")
```

```python
df["host_since_year"].unique()[0]
```

```python
from dash import Dash, dcc, html, Input, Output, callback

app = Dash(__name__)

app.layout = html.Div([
    dcc.Graph(id='graph-with-slider'),
    dcc.Slider(
        df['host_since_year'].min(),
        df['host_since_year'].max(),
        step=None,
        value=df['host_since_year'].min(),
        marks={str(host_since_year): str(host_since_year) for host_since_year in df['host_since_year'].unique()},
        id='host_since_year-slider'
    )
])


@callback(
    Output('graph-with-slider', 'figure'),
    Input('host_since_year-slider', 'value'))
def update_figure(selected_year):
    filtered_df = df[df.host_since_year == selected_year]

    fig = px.scatter(filtered_df, x="price", y="",
                     size="pop", color="room_type", hover_name="bathrooms",
                     log_x=True, size_max=30)

    fig.update_layout(transition_duration=500)

    return fig


if __name__ == '__main__':
    app.run(debug=True, port=8051)
```
